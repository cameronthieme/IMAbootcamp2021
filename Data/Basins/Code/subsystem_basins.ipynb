```python
def year_month(m,y):
    return 100*y + m
```

**Basin Amazonas**
* River Branco
* Jirau

```python
# River Branco
riobranco = pd.read_csv("../Amazonas/cidade_de_rio_branco_rio_acre.csv")
riobranco['year_month'] = year_month(riobranco['month'], riobranco['year'])
riobranco = riobranco[['year_month', 'accumulate']]
riobranco.rename({'accumulate':'accumulate_riobranco'}, axis = 1, inplace = True)

# Jirau
jirau = pd.read_csv("../Amazonas/uhe_jirau.csv")
jirau['year_month'] = year_month(jirau['month'], jirau['year'])
jirau= jirau[['year_month', 'accumulate']]
jirau.rename({'accumulate': 'accumulate_jirau'}, axis = 1, inplace = True)


#Merge all
Amazonas = pd.merge(riobranco, jirau, on =['year_month'])

#Total
Amazonas['Amazonas_total'] = Amazonas['accumulate_riobranco'] + Amazonas['accumulate_jirau'] 

Amazonas.head()
#Amazonas.to_csv("Amazonas/Amazonas_basin.csv")
```

**Basin Grande**
* River grande
* River pardo
* A. Vermelha
* Caconde
* Furnas
* Marimbondo
* Mascarenha
* Camargos

```python
# Rio Grande
riogrande = pd.read_csv("../Grande/incremental_rio_grande.csv")
riogrande['year_month'] = year_month(riogrande['month'], riogrande['year'])
riogrande.rename({'accumulate':'accumulate_riogrande'}, axis = 1, inplace = True)
riogrande = riogrande[['year_month', 'accumulate_riogrande']]

#River Pardo
pardo = pd.read_csv("../Grande/incremental_rio_pardo.csv")
pardo['year_month'] = year_month( pardo['month'], pardo['year'])
pardo= pardo[['year_month', 'accumulate']]
pardo.rename({'accumulate': 'accumulate_pardo'}, axis = 1, inplace = True)

# Reservoir Vermelha
vermelha = pd.read_csv("../Grande/incremental_uhe_agua_vermelha.csv")
vermelha['year_month'] = year_month(vermelha['month'], vermelha['year'])
vermelha.rename({'accumulate':'accumulate_vermelha'}, axis = 1, inplace = True)
vermelha = vermelha[['year_month', 'accumulate_vermelha']]

#Caconde
caconde = pd.read_csv("../Grande/incremental_uhe_caconde.csv")
caconde['year_month'] = year_month( caconde['month'], caconde['year'])
caconde= caconde[['year_month', 'accumulate']]
caconde.rename({'accumulate': 'accumulate_caconde'}, axis = 1, inplace = True)

# Reservoir Furnas
furnas = pd.read_csv("../Grande/incremental_uhe_furnas.csv")
furnas['year_month'] = year_month(furnas['month'], furnas['year'])
furnas.rename({'accumulate':'accumulate_furnas'}, axis = 1, inplace = True)
furnas = furnas[['year_month', 'accumulate_furnas']]

# Reservoir Marimbondo
marimbondo = pd.read_csv("../Grande/incremental_uhe_marimbondo.csv")
marimbondo['year_month'] = year_month(marimbondo['month'], marimbondo['year'])
marimbondo.rename({'accumulate':'accumulate_marimbondo'}, axis = 1, inplace = True)
marimbondo = marimbondo[['year_month', 'accumulate_marimbondo']]

#Mascarenha
mascarenha = pd.read_csv("../Grande/incremental_uhe_mascarenhas.csv")
mascarenha['year_month'] = year_month( mascarenha['month'], mascarenha['year'])
mascarenha= mascarenha[['year_month', 'accumulate']]
mascarenha.rename({'accumulate': 'accumulate_mascarenha'}, axis = 1, inplace = True)

#Camargos
camargos = pd.read_csv("../Grande/uhe_camargos.csv")
camargos['year_month'] = year_month( camargos['month'], camargos['year'])
camargos= camargos[['year_month', 'accumulate']]
camargos.rename({'accumulate': 'accumulate_camargos'}, axis = 1, inplace = True)

#Merge
Grande = pd.merge(riogrande, pardo, on = ['year_month'])
Grande = pd.merge(Grande, vermelha, on = ['year_month'])
Grande = pd.merge(Grande, caconde, on = ['year_month'])
Grande = pd.merge(Grande, furnas, on = ['year_month'])
Grande = pd.merge(Grande, marimbondo, on = ['year_month'])
Grande = pd.merge(Grande, mascarenha, on = ['year_month'])
Grande = pd.merge(Grande, camargos, on = ['year_month'])

#Total
Grande['Grande_total'] = Grande['accumulate_riogrande'] + Grande['accumulate_pardo'] + Grande['accumulate_vermelha'] + Grande['accumulate_caconde'] + Grande['accumulate_furnas'] + Grande['accumulate_marimbondo']+ Grande['accumulate_mascarenha'] + Grande['accumulate_camargos']

Grande.head(10)
#Grande.to_csv("Grande/Grade_basin.csv")
```

```python
#River parainaba: basin
basinp = pd.read_csv("../Paranaiba/incremental_bacia_do_rio_paranaiba.csv")
basinp['year_month'] = year_month(basinp['month'], basinp['year'])
basinp.rename({'accumulate':'accumulate_basinp'}, axis = 1, inplace = True)
basinp = basinp[['year_month', 'accumulate_basinp']]

#River Parnaiba
riop = pd.read_csv("../Paranaiba/incremental_rio_parnaiba.csv")
riop['year_month'] = year_month(riop['month'], riop['year'])
riop.rename({'accumulate':'accumulate_riop'}, axis = 1, inplace = True)
riop = riop[['year_month', 'accumulate_riop']]

# Corumba 
corumba = pd.read_csv("../Paranaiba/incremental_rio_corumba.csv")
corumba['year_month'] = year_month(corumba['month'], corumba['year'])
corumba= corumba[['year_month', 'accumulate']]
corumba.rename({'accumulate': 'accumulate_corumba'}, axis = 1, inplace = True)

# Corumba 1
corumba1 = pd.read_csv("../Paranaiba/incremental_uhe_corumba_i.csv")
corumba1['year_month'] = year_month(corumba1['month'], corumba1['year'])
corumba1= corumba1[['year_month', 'accumulate']]
corumba1.rename({'accumulate': 'accumulate_corumba1'}, axis = 1, inplace = True)

# Corumba 3
corumba3 = pd.read_csv("../Paranaiba/incremental_uhe_corumba_iii.csv")
corumba3['year_month'] = year_month(corumba3['month'], corumba3['year'])
corumba3= corumba3[['year_month', 'accumulate']]
corumba3.rename({'accumulate': 'accumulate_corumba3'}, axis = 1, inplace = True)

# Corumba 4
corumba4 = pd.read_csv("../Paranaiba/incremental_uhe_corumba_iv.csv")
corumba4['year_month'] = year_month(corumba4['month'], corumba4['year'])
corumba4= corumba4[['year_month', 'accumulate']]
corumba4.rename({'accumulate': 'accumulate_corumba4'}, axis = 1, inplace = True)

# Reservoir Sao Simao
simao = pd.read_csv("../Paranaiba/incremental_uhe_sao_simao.csv")
simao['year_month'] = year_month(simao['month'], simao['year'])
simao.rename({'accumulate':'accumulate_simao'}, axis = 1, inplace = True)
simao = simao[['year_month', 'accumulate_simao']]

# Reservoir Emborcacao
emborcacao = pd.read_csv("../Paranaiba/uhe_emborcacao.csv")
emborcacao['year_month'] = year_month(emborcacao['month'], emborcacao['year'])
emborcacao.rename({'accumulate':'accumulate_emborcacao'}, axis = 1, inplace = True)
emborcacao = emborcacao[['year_month', 'accumulate_emborcacao']]

# Reservoir Itumbiara
itumbiara = pd.read_csv("../Paranaiba/uhe_itumbiara.csv")
itumbiara['year_month'] = year_month(itumbiara['month'], itumbiara['year'])
itumbiara.rename({'accumulate':'accumulate_itumbiara'}, axis = 1, inplace = True)
itumbiara = itumbiara[['year_month', 'accumulate_itumbiara']]

#Merge
Paranaiba = pd.merge(basinp, riop, on = ['year_month'])
Paranaiba = pd.merge(Paranaiba, simao, on = ['year_month'])
Paranaiba = pd.merge(Paranaiba, emborcacao, on = ['year_month'])
Paranaiba = pd.merge(Paranaiba, itumbiara, on = ['year_month'])
Paranaiba = pd.merge(Paranaiba, corumba, on = ['year_month'])
Paranaiba = pd.merge(Paranaiba, corumba1, on = ['year_month'])
Paranaiba = pd.merge(Paranaiba, corumba3, on = ['year_month'])
Paranaiba = pd.merge(Paranaiba, corumba4, on = ['year_month'])

Paranaiba['Paranaiba_total'] = Paranaiba['accumulate_basinp'] + Paranaiba['accumulate_riop'] + Paranaiba['accumulate_simao'] + Paranaiba['accumulate_emborcacao'] + Paranaiba['accumulate_itumbiara'] + Paranaiba['accumulate_corumba'] +Paranaiba ['accumulate_corumba1'] +Paranaiba['accumulate_corumba3'] + Paranaiba['accumulate_corumba4']

Paranaiba.head(10)
#Paranaiba.to_csv("Paranaiba/Paranaiba_basin.csv")
```

**Doce**
* River Doce
* Baguari
* River Picacicaba
* River Santo Antonio

```python
# River Doce
riodoce = pd.read_csv("../Doce/incremental_rio_doce.csv")
riodoce['year_month'] = year_month(riodoce['month'], riodoce['year'])
riodoce.rename({'accumulate':'accumulate_riodoce'}, axis = 1, inplace = True)
riodoce = riodoce[['year_month', 'accumulate_riodoce']]

# Baguari
baguari = pd.read_csv("../Doce/incremental_uhe_baguari.csv")
baguari['year_month'] = year_month(baguari['month'], baguari['year'])
baguari.rename({'accumulate':'accumulate_baguari'}, axis = 1, inplace = True)
baguari = baguari[['year_month', 'accumulate_baguari']]

# River Piracicaba
piracicaba = pd.read_csv("../Doce/rio_piracicaba.csv")
piracicaba['year_month'] = year_month(piracicaba['month'], piracicaba['year'])
piracicaba.rename({'accumulate':'accumulate_piracicaba'}, axis = 1, inplace = True)
piracicaba = piracicaba[['year_month', 'accumulate_piracicaba']]

# River Santo Antonio
antonio = pd.read_csv("../Doce/rio_santo_antonio.csv")
antonio['year_month'] = year_month(antonio['month'], antonio['year'])
antonio.rename({'accumulate':'accumulate_antonio'}, axis = 1, inplace = True)
antonio = antonio[['year_month', 'accumulate_antonio']]

#Merge
Doce = pd.merge(riodoce, baguari, on=['year_month'])
Doce = pd.merge(Doce, piracicaba, on=['year_month'])
Doce = pd.merge(Doce, antonio, on=['year_month'])

#Total
Doce['Doce_total']= Doce['accumulate_riodoce']+Doce['accumulate_baguari']+ Doce['accumulate_piracicaba'] + Doce['accumulate_antonio']

Doce.head(10)
#Doce.to_csv("Doce/Doce_basin.csv")
```

```python
#Final Merge

DF = pd.merge(Grande, Paraiba_do_sul, on = ['year_month'])
DF = pd.merge(DF, Parana, on = ['year_month'])
DF = pd.merge(DF, Paranaiba, on = ['year_month'])
DF = pd.merge(DF, Paranapanema, on = ['year_month'])
DF = pd.merge(DF, SaoFrancisco, on = ['year_month'])
DF = pd.merge(DF, Tiete, on = ['year_month'])
DF = pd.merge(DF, Tocantins, on = ['year_month'])
DF = pd.merge(DF, Amazonas, on = ['year_month'])
DF = pd.merge(DF, Doce, on = ['year_month'])

DF = DF[['year', 'month', 'Amazonas_total', 'Grande_total', 'Paraiba_do_sul_total', 'Parana_total', 'Paranaiba_total', 'Paranapanema_total', 'SaoFrancisco_total', "Tiete_total", "Tocantins_total", "Doce_total"]]
DF['day'] = 1
df = pd.DataFrame({'year' : DF['year'],
                  'month' : DF['month'],
                  'day' : DF['day']})

DF['date'] = pd.to_datetime(df)
DF.index = DF['date']

DF['Total_accumulate'] = DF['Amazonas_total']+ DF['Grande_total'] + DF['Paraiba_do_sul_total'] + DF['Parana_total'] + DF['Paranaiba_total'] + DF['Paranapanema_total'] + DF['SaoFrancisco_total'] + DF["Tiete_total"] + DF["Tocantins_total"] + DF["Doce_total"]

DF = DF[['date', 'year', 'month', 'Total_accumulate']]
DF.head(10)
DF.to_csv("../accumulate_rain_subsystem.csv")
```
