**Basin Amazonas**
* River Branco
* River Acre
* River Madeira
* River Xingu
* Belo Monte
* Jirau

```python
# River Branco
riobranco = pd.read_csv("../Amazonas/cidade_de_rio_branco_rio_acre.csv")
riobranco['year_month'] = year_month(riobranco['month'], riobranco['year'])
riobranco = riobranco[['year_month', 'accumulate']]
riobranco.rename({'accumulate':'accumulate_riobranco'}, axis = 1, inplace = True)

# River Acre
rioacre = pd.read_csv("../Amazonas/incremental_rio_acre.csv")
rioacre['year_month'] = year_month(rioacre['month'], rioacre['year'])
rioacre = rioacre[['year_month', 'accumulate']]
rioacre.rename({'accumulate':'accumulate_rioacre'}, axis = 1, inplace = True)

# River Madeira
madeira = pd.read_csv("../Amazonas/incremental_rio_madeira.csv")
madeira['year_month'] = year_month( madeira['month'], madeira['year'])
madeira= madeira[['year_month', 'accumulate']]
madeira.rename({'accumulate': 'accumulate_madeira'}, axis = 1, inplace = True)

# River Xingu
xingu = pd.read_csv("../Amazonas/incremental_rio_xingu.csv")
xingu['year_month'] = year_month( xingu['month'], xingu['year'])
xingu= xingu[['year_month', 'accumulate']]
xingu.rename({'accumulate': 'accumulate_xingu'}, axis = 1, inplace = True)

# Belo Monte
monte = pd.read_csv("../Amazonas/uhe_belo_monte.csv")
monte['year_month'] = year_month( monte['month'], monte['year'])
monte= monte[['year_month', 'accumulate']]
monte.rename({'accumulate': 'accumulate_monte'}, axis = 1, inplace = True)

# Jirau
jirau = pd.read_csv("../Amazonas/uhe_jirau.csv")
jirau['year_month'] = year_month(jirau['month'], jirau['year'])
jirau= jirau[['year_month', 'accumulate']]
jirau.rename({'accumulate': 'accumulate_jirau'}, axis = 1, inplace = True)


#Merge all
Amazonas = pd.merge(riobranco, rioacre, on =['year_month'])
Amazonas = pd.merge(Amazonas, madeira , on=['year_month'])
Amazonas = pd.merge(Amazonas, xingu , on=['year_month'])
Amazonas = pd.merge(Amazonas, monte, on=['year_month'])
Amazonas = pd.merge(Amazonas, jirau, on=['year_month'])

#Total
Amazonas['Amazonas_total'] = Amazonas['accumulate_riobranco'] + Amazonas['accumulate_rioacre'] + Amazonas['accumulate_madeira'] + Amazonas['accumulate_xingu'] + Amazonas['accumulate_monte'] + Amazonas['accumulate_jirau'] 

Amazonas.head()
Amazonas.to_csv("../Amazonas/Amazonas_basin.csv")
```

```python
# River Doce
riodoce = pd.read_csv("../Doce/incremental_rio_doce.csv")
riodoce['year_month'] = year_month(riodoce['month'], riodoce['year'])
riodoce.rename({'accumulate':'accumulate_riodoce'}, axis = 1, inplace = True)
riodoce = riodoce[['year_month', 'accumulate_riodoce']]

# Baguari
baguari = pd.read_csv("../Doce/incremental_uhe_baguari.csv")
baguari['year_month'] = year_month(baguari['month'], baguari['year'])
baguari.rename({'accumulate':'accumulate_baguari'}, axis = 1, inplace = True)
baguari = baguari[['year_month', 'accumulate_baguari']]

# River Piracicaba
piracicaba = pd.read_csv("../Doce/rio_piracicaba.csv")
piracicaba['year_month'] = year_month(piracicaba['month'], piracicaba['year'])
piracicaba.rename({'accumulate':'accumulate_piracicaba'}, axis = 1, inplace = True)
piracicaba = piracicaba[['year_month', 'accumulate_piracicaba']]

# River Santo Antonio
antonio = pd.read_csv("../Doce/rio_santo_antonio.csv")
antonio['year_month'] = year_month(antonio['month'], antonio['year'])
antonio.rename({'accumulate':'accumulate_antonio'}, axis = 1, inplace = True)
antonio = antonio[['year_month', 'accumulate_antonio']]

#Merge
Doce = pd.merge(riodoce, baguari, on=['year_month'])
Doce = pd.merge(Doce, piracicaba, on=['year_month'])
Doce = pd.merge(Doce, antonio, on=['year_month'])

#Total
Doce['Doce_total']= Doce['accumulate_riodoce']+Doce['accumulate_baguari']+ Doce['accumulate_piracicaba'] + Doce['accumulate_antonio']

Doce.head(10)
Doce.to_csv("../Doce/Doce_basin.csv")
```
