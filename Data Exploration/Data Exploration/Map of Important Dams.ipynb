```python
def create_map_rivers():
    #y_low,y_up,x_low,x_up=(-33.8689056,5.2842873,-73.9830625,-28.6341164)
    y_low,y_up,x_low,x_up=(-28,-10,-73.9830625,-35)
    m = Basemap(llcrnrlon=x_low,llcrnrlat=y_low,
                urcrnrlon=x_up,urcrnrlat=y_up, epsg=29101,resolution='h',)
    #m.arcgisimage(service='ESRI_Imagery_World_2D', xpixels = 1500, verbose= True)
    m.shadedrelief()
    m.drawrivers(linewidth=2,color='deepskyblue')
    m.drawcountries(linewidth=3,linestyle='--',color='k')
    return m
```
