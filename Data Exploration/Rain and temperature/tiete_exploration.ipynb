## *Import packages*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import keras as keras
import plotly.express as px
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import interactive
interactive(True)
```

# *Import csv file*

```python
month_tiete = pd.read_csv("../../Data/month_tiete.csv")
month_tiete.head(10)
```

## *Understanding accumulate*

```python
sns.set_theme(style="whitegrid")
sns.lineplot(data = month_tiete, x = 'count', y= 'accumulate', palette = 'Spectral')
```

```python
sns.set_theme(style="whitegrid")
sns.lineplot(data = month_tiete, x = 'month', y= 'accumulate', hue = 'year', palette = 'Spectral')
```

```python
sns.set_theme(style="ticks")

g = sns.jointplot(x="accumulate", y="total_rain", data=month_tiete, hue = "year", palette = "viridis")
```

```python
sns.set_theme(style="ticks")

g = sns.jointplot(x="accumulate", y="energy_generated", data=month_tiete, hue = "year", palette = "rocket")
```

```python
sns.set_theme(style="ticks")

g = sns.jointplot(x="accumulate", y="effluent_flow", data=month_tiete, hue = "year", palette = "mako")
```

```python
sns.set_theme(style="ticks")

g = sns.jointplot(x="accumulate", y="influent_flow", data=month_tiete, hue = "year", palette = "Spectral")
```
